```python
#Display sales total by state  and year

total_sales_2017_by_state = sales_data_df.groupby('StateName')[['1/31/2017', '2/28/2017', '3/31/2017', '4/30/2017', '5/31/2017', '6/30/2017', '7/31/2017', '8/31/2017', '9/30/2017', '10/31/2017', '11/30/2017', '12/31/2017']].sum()

total_sales_2018_by_state= sales_data_df.groupby('StateName')[['1/31/2018', '2/28/2018', '3/31/2018', '4/30/2018', '5/31/2018', '6/30/2018', '7/31/2018', '8/31/2018', '9/30/2018', '10/31/2018', '11/30/2018', '12/31/2018']].sum()

total_sales_2019_by_state = sales_data_df.groupby('StateName')[['1/31/2019', '2/28/2019', '3/31/2019', '4/30/2019', '5/31/2019', '6/30/2019', '7/31/2019', '8/31/2019', '9/30/2019', '10/31/2019', '11/30/2019', '12/31/2019']].sum()

total_sales_2020_by_state =  sales_data_df.groupby('StateName')[['1/31/2020', '2/29/2020', '3/31/2020', '4/30/2020', '5/31/2020', '6/30/2020', '7/31/2020', '8/31/2020', '9/30/2020', '10/31/2020', '11/30/2020', '12/31/2020']].sum()

total_sales_2021_by_state = sales_data_df.groupby('StateName')[['1/31/2021', '2/28/2021', '3/31/2021', '4/30/2021', '5/31/2021', '6/30/2021', '7/31/2021', '8/31/2021', '9/30/2021', '10/31/2021', '11/30/2021', '12/31/2021']].sum()

total_sales_2022_by_state = sales_data_df.groupby('StateName')[['1/31/2022', '2/28/2022', '3/31/2022', '4/30/2022', '5/31/2022', '6/30/2022', '7/31/2022', '8/31/2022', '9/30/2022', '10/31/2022', '11/30/2022', '12/31/2022']].sum()

total_sales_2023_by_state = sales_data_df.groupby('StateName')[['1/31/2023', '2/28/2023', '3/31/2023', '4/30/2023', '5/31/2023', '6/30/2023']].sum()

#find the average 

total_sales_2017_by_state['average total 2017'] = total_sales_2017_by_state.sum(axis=1) / 12
total_sales_2018_by_state['average total 2018'] = total_sales_2018_by_state.sum(axis=1) / 12
total_sales_2019_by_state['average total 2019'] = total_sales_2019_by_state.sum(axis=1) / 12
total_sales_2020_by_state['average total 2020'] = total_sales_2020_by_state.sum(axis=1) / 12
total_sales_2021_by_state['average total 2021'] = total_sales_2021_by_state.sum(axis=1) / 12
total_sales_2022_by_state['average total 2022'] = total_sales_2022_by_state.sum(axis=1) / 12
total_sales_2023_by_state['average total 2023'] = total_sales_2023_by_state.sum(axis=1) / 6
```

```python
#Find the data
winter_sales_2019_df = sales_data_df.groupby('StateName')[['1/31/2019', '2/28/2019','12/31/2019']].sum()

winter_sales_2020_df =  sales_data_df.groupby('StateName')[['1/31/2020', '2/29/2020','12/31/2020']].sum()

winter_sales_2021_df = sales_data_df.groupby('StateName')[['1/31/2021', '2/28/2021','12/31/2021']].sum()

winter_sales_2022_df = sales_data_df.groupby('StateName')[['1/31/2022', '2/28/2022','12/31/2022']].sum()


spring_sales_2019_df = sales_data_df.groupby('StateName')[[ '3/31/2019', '4/30/2019', '5/31/2019']].sum()

spring_sales_2020_df =  sales_data_df.groupby('StateName')[['3/31/2020', '4/30/2020', '5/31/2020']].sum()

spring_sales_2021_df = sales_data_df.groupby('StateName')[['3/31/2021', '4/30/2021', '5/31/2021']].sum()

spring_sales_2022_df = sales_data_df.groupby('StateName')[['3/31/2022', '4/30/2022', '5/31/2022']].sum()


summer_sales_2019_df = sales_data_df.groupby('StateName')[['6/30/2019', '7/31/2019', '8/31/2019']].sum()

summer_sales_2020_df =  sales_data_df.groupby('StateName')[['6/30/2020', '7/31/2020', '8/31/2020']].sum()

summer_sales_2021_df = sales_data_df.groupby('StateName')[['6/30/2021', '7/31/2021', '8/31/2021']].sum()

summer_sales_2022_df = sales_data_df.groupby('StateName')[['6/30/2022', '7/31/2022', '8/31/2022']].sum()


fall_sales_2019_df = sales_data_df.groupby('StateName')[['9/30/2019', '10/31/2019', '11/30/2019']].sum()

fall_sales_2020_df =  sales_data_df.groupby('StateName')[['9/30/2020', '10/31/2020', '11/30/2020']].sum()

fall_sales_2021_df = sales_data_df.groupby('StateName')[['9/30/2021', '10/31/2021', '11/30/2021']].sum()

fall_sales_2022_df = sales_data_df.groupby('StateName')[['9/30/2022', '10/31/2022', '11/30/2022']].sum()



#SFind averega sales per season and year

average_winter_2019_df = pd.DataFrame(winter_sales_2019_df.sum(axis=1) / 3, columns=['Winter_2019'])
average_winter_2020_df = pd.DataFrame(winter_sales_2020_df.sum(axis=1) / 3, columns=['Winter_2020'])
average_winter_2021_df = pd.DataFrame(winter_sales_2021_df.sum(axis=1) / 3, columns=['Winter_2021'])
average_winter_2022_df = pd.DataFrame(winter_sales_2022_df.sum(axis=1) / 3, columns=['Winter_2022'])



average_spring_2019_df = pd.DataFrame(spring_sales_2019_df.sum(axis=1) / 3, columns=['Spring_2019'])
average_spring_2020_df = pd.DataFrame(spring_sales_2020_df.sum(axis=1) / 3, columns=['Spring_2020'])
average_spring_2021_df = pd.DataFrame(spring_sales_2021_df.sum(axis=1) / 3, columns=['Spring_2021'])
average_spring_2022_df = pd.DataFrame(spring_sales_2022_df.sum(axis=1) / 3, columns=['Spring_2022'])


average_summer_2019_df = pd.DataFrame(summer_sales_2019_df.sum(axis=1) / 3, columns=['Summer_2019'])
average_summer_2020_df = pd.DataFrame(summer_sales_2020_df.sum(axis=1) / 3, columns=['Summer_2020'])
average_summer_2021_df = pd.DataFrame(summer_sales_2021_df.sum(axis=1) / 3, columns=['Summer_2021'])
average_summer_2022_df = pd.DataFrame(summer_sales_2022_df.sum(axis=1) / 3, columns=['Summer_2022'])




average_fall_2019_df = pd.DataFrame(fall_sales_2019_df.sum(axis=1) / 3, columns=['Fall_2019'])
average_fall_2020_df = pd.DataFrame(fall_sales_2020_df.sum(axis=1) / 3, columns=['Fall_2020'])
average_fall_2021_df = pd.DataFrame(fall_sales_2021_df.sum(axis=1) / 3, columns=['Fall_2021'])
average_fall_2022_df = pd.DataFrame(fall_sales_2022_df.sum(axis=1) / 3, columns=['Fall_2022'])


```

```python
#Create DF to input all data
seasonal_sales_df = pd.concat([
    average_winter_2019_df, average_winter_2020_df, average_winter_2021_df, average_winter_2022_df,
    average_spring_2019_df, average_spring_2020_df, average_spring_2021_df, average_spring_2022_df,
    average_summer_2019_df, average_summer_2020_df, average_summer_2021_df, average_summer_2022_df,
    average_fall_2019_df, average_fall_2020_df, average_fall_2021_df, average_fall_2022_df
], axis=1)


seasonal_sales_df
```
